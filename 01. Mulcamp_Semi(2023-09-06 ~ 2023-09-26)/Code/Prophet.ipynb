## Classify APT, Officetel, and Townhouse

```python
data_APT = data[data['Building_Use'].str.contains('아파트')]
data_APT
```

```python
data_officetel = data[data['Building_Use'].str.contains('오피스텔')]
data_officetel
```

```python
data_townhouse = data[data['Building_Use'].str.contains('연립다세대')]
data_townhouse
```

## Extract 50000 samples from the total APT data

```python
# Randomly select 50,000 rows of the total data_APT
sample_size = 50000
data_APT_50000 = data_APT.sample(n=sample_size, random_state=42)
data_APT_50000
```

## Sort the 50000 APT data in ascending order

```python
# Sorting by the columns in descending order
data_APT_50000 = data_APT_50000.sort_values(by=['YearMonth'], ascending=[True])

data_APT_50000
```

## Convert the date type for the Prophet tool.

```python
data_APT_50000['YearMonth'] = pd.to_datetime(data_APT_50000['YearMonth'], format='%Y%m')
data_officetel['YearMonth'] = pd.to_datetime(data_officetel['YearMonth'], format='%Y%m')
data_townhouse['YearMonth'] = pd.to_datetime(data_townhouse['YearMonth'], format='%Y%m')
```

## Preprocess the data using drop( )

```python
columns_to_drop = ['위도', '경도', 'Region_Name', 'Building_Use']
data_APT_50000 = data_APT_50000.drop(columns = columns_to_drop)
data_APT_50000
```

```python
columns_to_drop = ['위도', '경도', 'Region_Name', 'Building_Use']
data_officetel = data_officetel.drop(columns = columns_to_drop)
data_officetel
```

```python
columns_to_drop = ['위도', '경도', 'Region_Name', 'Building_Use']
data_townhouse = data_townhouse.drop(columns = columns_to_drop)
data_townhouse
```

## Modeling of the 50000 APT data by Prophet

```python
import pandas as pd
from prophet import Prophet

# Rename columns for easier usage in Prophet
data_APT_50000.rename(columns={"YearMonth": "ds", "JS_Price": "y"}, inplace=True)


# Create an instance of the Prophet model
model = Prophet()

# Fit the model on your data
model.fit(data_APT_50000)

# Make predictions with the trained model
future_dates_APT = model.make_future_dataframe(periods=365*3)

forecast_APT = model.predict(future_dates_APT)

# Access the predicted future prices
future_prices_APT = forecast_APT[['ds', 'yhat']]

# Convert date column to desired output format ('YYYY-MM-DD')
future_prices_APT['ds'] = future_prices_APT['ds'].dt.strftime('%Y-%m-%d')

# Rename columns for clarity
future_prices_APT.rename(columns={'ds': 'Date', 'yhat': 'Predicted_Price'}, inplace=True)

future_prices_APT
```

## Convert the date format properly

```python
# Convert the "Date" column to datetime if it's not already in datetime format
future_prices_APT['Date'] = pd.to_datetime(future_prices_APT['Date'])

# Extract rows where the day is 01 (1st day of the month)
future_prices_APT = future_prices_APT[future_prices_APT['Date'].dt.day == 1]
future_prices_APT
```

## APT Plot

```python
import matplotlib.pyplot as plt

# Let's assume you already have a DataFrame named 'future_prices' with 'Date' and 'Predicted_Price' columns.

# Plotting with Matplotlib
plt.figure(figsize=(120, 60))
plt.plot(future_prices_APT['Date'], future_prices_APT['Predicted_Price'], linewidth=20)
plt.xlabel('Date')
plt.ylabel('Predicted Price')
plt.title('APT - Predicted Prices Over Time')

# Set the x-axis ticks for monthly intervals
monthly_ticks = pd.date_range(start=future_prices_APT['Date'].min(), end=future_prices_APT['Date'].max(), freq='MS')
plt.xticks(monthly_ticks, [date.strftime('%Y-%m-%d') for date in monthly_ticks], rotation=45)

plt.grid(True)
plt.show()
```
