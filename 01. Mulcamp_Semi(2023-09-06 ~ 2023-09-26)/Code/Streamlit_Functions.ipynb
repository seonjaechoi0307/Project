# LightGBM Imputation Test (Population, IR)

```python
import pandas as pd
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np

# Load your dataset


# Step 1: Split the data into subsets with and without missing values for both 'Population' and 'IR'
data_with_missing = data[data['Population'].isnull() | data['IR'].isnull()]
data_without_missing = data.dropna(subset=['Population', 'IR'])

# Check if there are missing values to impute
if not data_with_missing.empty:
    # Step 2: Train separate LightGBM models for each target column
    for target_column in ['Population', 'IR']:
        X = data_without_missing.drop(columns=[target_column])
        y = data_without_missing[target_column]

        lgb_model = lgb.LGBMRegressor(n_estimators=1000, random_state=142)
        lgb_model.fit(X, y)

        # Step 3: Use the trained model to predict missing values for each target column
        imputed_values = lgb_model.predict(data_with_missing.drop(columns=[target_column]))

        # Fill in missing values in the original dataset for the specific target column
        data.loc[data_with_missing.index, target_column] = imputed_values

        # Step 4: Split the dataset for evaluation (optional)
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=234)

        # Step 5: Calculate RMSE on the test set to evaluate the model (optional)
        y_true = data_with_missing[target_column]  # True values of the missing column
        y_pred = imputed_values  # Predicted values
        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        print(f"RMSE for {target_column}: {rmse}")

else:
    print("No missing values to impute in the columns")

# Assuming 'data' is the DataFrame you want to inspect at the end
data.info()
data
```
