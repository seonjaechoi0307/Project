```python
import pandas as pd

data = pd.read_csv('/content/한국은행_기준금리(월별).csv', encoding='euc-kr')


# Create a DataFrame
df = pd.DataFrame(data)

# Split "통계표" into separate columns for Month and Year
df[['Month', 'Year']] = df['통계표'].str.split('-', expand=True)

# Drop the original "통계표" column
df = df.drop(columns=['통계표'])

# Display the DataFrame with Month and Year columns
print(df)
```

```python
# Define a dictionary to map English month names to numbers
month_mapping = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
    'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}

# Map the Month column to numerical values
df['Month'] = df['Month'].map(month_mapping)

# Display the DataFrame with Month and Year columns
print(df)
```

```python
# Convert two-digit years to four-digit years
df['Year'] = '20' + df['Year']

print(df)
```

```python
df = df.rename(columns={'1.3.1. 한국은행 기준금리 및 여수신금리': 'IR'})
df
```

```python
df.to_csv('한국은행_기준금리(월별)_converted.csv', index=False)
```

```python
import pandas as pd

total_df = pd.read_csv('/content/Total_Data.csv', encoding='utf-8')

crime_df = pd.read_csv('/content/서울시5대범죄전처리완료.csv', encoding='euc-kr')

crime_df = crime_df.rename(columns={'rates': 'Crime_Rates'})

Total_Data = total_df.merge(crime_df[['Year', 'Region_Name', 'Crime_Rates']], on=['Year', 'Region_Name'], how='left')

print(Total_Data)
```

```python
Total_Data.to_csv('Total_Data.csv', index=False)
```

## Preprocessing of missing values using Regression Imputation

```python
import pandas as pd
from sklearn.linear_model import LinearRegression


# Create a subset of data with non-missing values in predictor and target variables
train_data = Total_Data.dropna(subset=['Year', 'Crime_Rates'])

# Separate predictor and target variables in the training set
X_train = train_data[['Year']]
y_train = train_data['Crime_Rates']

# Fit a regression model on the training data
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)

# Predict missing values using the trained model
X_missing = Total_Data.loc[Total_Data['Crime_Rates'].isnull(), ['Year']]
predicted_values = regression_model.predict(X_missing)

# Fill in missing values with predicted values
Total_Data.loc[Total_Data['Crime_Rates'].isnull(), 'Crime_Rates'] = predicted_values
```

```python
Total_Data.to_csv('Total_Data.csv', index=False)
```
