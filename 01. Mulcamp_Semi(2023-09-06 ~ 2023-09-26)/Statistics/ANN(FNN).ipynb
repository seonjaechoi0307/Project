## Modified Z-score for Outliers

```python
import numpy as np

# Assuming 'data' is your DataFrame containing the dataset

data_APT = data_APT.reset_index(drop=True)


data = data_APT

def modified_z_score(data):
    median = np.median(data)
    median_absolute_deviation = np.median(np.abs(data - median))

    modified_z_scores = 0.6745 * (data - median) / median_absolute_deviation

    return modified_z_scores

# Extract the 'Sell_Price' column for outlier detection
sell_price = data['Sell_Price']

# Calculate modified z-scores for Sell_Price
modified_z_scores = modified_z_score(sell_price)

# Set a threshold value for outlier detection
threshold = 3.5

# Find indices of outliers above the threshold
outlier_indices = np.where(modified_z_scores > threshold)[0]

# Delete rows with outliers from the DataFrame
data_APT = data.drop(outlier_indices)


# Number and percentage of outliers
print(f"Number of outliers: {len(outlier_indices)}")

percentage_outliers = (len(outlier_indices) / len(data)) * 100

print(f"Percentage of outliers: {percentage_outliers:.2f}%")

data_APT
```

## Calculate the shortest distances between each real estate property and features

#### APT - Features

```python
import numpy as np
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

# Load real estate and subway station data
real_estate_data = data_APT  # Replace with your real estate data file
subway_station_data = pd.read_csv('/content/drive/MyDrive/Coordinates_Merge/Subway_Stations.csv', encoding = 'utf-8')  # Replace with your subway station data file

# Function to calculate Haversine distance
def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371000  # Radius of Earth in meters
    phi1 = radians(lat1)
    phi2 = radians(lat2)

    delta_phi = radians(lat2 - lat1)
    delta_lambda = radians(lon2 - lon1)

    a = sin(delta_phi / 2) * sin(delta_phi / 2) + cos(phi1) * cos(phi2) * sin(delta_lambda / 2) * sin(delta_lambda / 2)
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c  # Distance in meters
    return distance

# Calculate shortest distance for each real estate to each subway station
shortest_distances = []

for _, real_estate_row in real_estate_data.iterrows():
    real_estate_lat = real_estate_row['위도']
    real_estate_lon = real_estate_row['경도']

    distances_to_subway_stations = []
    for _, subway_station_row in subway_station_data.iterrows():
        subway_station_lat = subway_station_row['위도']
        subway_station_lon = subway_station_row['경도']

        distance = haversine_distance(real_estate_lat, real_estate_lon, subway_station_lat, subway_station_lon)
        distances_to_subway_stations.append(distance)

    shortest_distance = min(distances_to_subway_stations)
    shortest_distances.append(shortest_distance)

# Add the shortest distances to subway stations to the real estate data
real_estate_data['Shortest_Distance_to_Subway'] = shortest_distances

# Save the merged data to a new CSV file
real_estate_data.to_csv('data_APT.csv', index=False)

data_APT = real_estate_data

data_APT
```
